```python
from io import BytesIO
import requests
from PIL import Image
import matplotlib.pyplot as plt

# @title cargar imagen
url1 = "https://raw.githubusercontent.com/semilleroCV/Hands-on-Computer-Vision/main/sesiones/sesion1/images/perro.jpg"
url2 = "https://assets.elanco.com/8e0bf1c2-1ae4-001f-9257-f2be3c683fb1/fca42f04-2474-4302-a238-990c8aebfe8c/Siamese_cat_1110x740.jpg?w=3840&q=75&auto=format"

img1 = None
img2 = None

# Process url1
response1 = requests.get(url1)
if response1.status_code == 200:
    content_type1 = response1.headers.get('Content-Type', '')
    if 'image' in content_type1:
        try:
            img1 = Image.open(BytesIO(response1.content))
        except Image.UnidentifiedImageError:
            print(f"Error: PIL could not identify image from {url1}. Content-Type: {content_type1}")
    else:
        print(f"Error: Content from {url1} is not an image. Content-Type: {content_type1}")
else:
    print(f"Error: Failed to fetch image from {url1}. Status Code: {response1.status_code}")

# Process url2
response2 = requests.get(url2)
if response2.status_code == 200:
    content_type2 = response2.headers.get('Content-Type', '')
    if 'image' in content_type2:
        try:
            img2 = Image.open(BytesIO(response2.content))
        except Image.UnidentifiedImageError:
            print(f"Error: PIL could not identify image from {url2}. Content-Type: {content_type2}")
    else:
        print(f"Error: Content from {url2} is not an image. Content-Type: {content_type2}")
else:
    print(f"Error: Failed to fetch image from {url2}. Status Code: {response2.status_code}")

plt.figure(figsize=(10, 5))

if img1:
    plt.subplot(1, 2, 1)
    plt.imshow(img1)
    plt.title("Image 1")
    plt.axis("off")
else:
    plt.subplot(1, 2, 1)
    plt.text(0.5, 0.5, "Image 1 failed to load", horizontalalignment='center', verticalalignment='center', transform=plt.gca().transAxes)
    plt.title("Image 1")
    plt.axis("off")

if img2:
    plt.subplot(1, 2, 2)
    plt.imshow(img2)
    plt.title("Image 2")
    plt.axis("off")
else:
    plt.subplot(1, 2, 2)
    plt.text(0.5, 0.5, "Image 2 failed to load", horizontalalignment='center', verticalalignment='center', transform=plt.gca().transAxes)
    plt.title("Image 2")
    plt.axis("off")

plt.show()
```

```python

# @title clasificar imagen
classifier = pipeline('image-classification')
result1 = classifier(url1)
result2 = classifier(url2)

print(f"\n **La imagen es un: {result1[0]['label']}** ")
print(f"\n **La imagen es un: {result2[0]['label']}** ")
```
